*Ejercicio:* Definir una función que decida si un número es perfecto. Un número natural cumple esta propiedad cuando es igual a la suma de sus divisores (excepto el propio número).

```python
def esPerfecto(n):
    sumatorio = 0 
    limite = n//2+1
    for i in range(1,limite): 
        if n%i == 0: 
            sumatorio += i 
    return n == sumatorio
print(esPerfecto(28))
print(esPerfecto(20))
```

**Dictionary** - Un diccionario (dict) es un conjunto de pares clave:valor, que puede ser usado para construir diferentes estructuras (p.ej. una tabla de registros). Es una estructura mutable.

```python
eng2sp = {'one':'uno', 'two':'dos', 'three':'tres'}
print(eng2sp['two']) # ojo: son parentesis cuadrados
```

*Ejercicio:* Usar un diccionario para crear una una secuencia de con las letras de una cadena (string) seguidas por su frecuencia de aparición en la misma. Resolver el ejercicio haciendo uso de una función.

```python
# funcion que calcula la frecuencia de las letras de una palabra
def frec(palabra):
    d = dict()
    for letra in palabra: 
        if letra not in d: 
            d[letra] = 1 
        else: 
            d[letra] += 1 
    return d
print(frec('brontosaurio'))
```

### Ejemplos de uso de NumPy

Se presentan ejemplos sencillos de creación de vectores, matrices, obtención de sus dimensiones de un array, creación de arrays constantes, etc. Nótese que los vectores se representan como listas y las matrices como listas de listas.

```python
import numpy as np

a = np.array([1, 2, 3])   # Crea un vector
print(type(a))            # Imprime su tipo
print(a.shape)            # Imprime la dimensión
print(a[0], a[2])         # Imprime algunos elementos del vector
a[0] = 5                  # Modifica un elemento del vector
print(a)                  # Imprime el vector completo

b = np.array([[1,2,3],[4,5,6]])    # Crea una matriz bidimensional
print(b.shape)                     # Imprime sus dimensiones
print(b[0, 0], b[0, 1], b[1, 2])   # Imprime algunos elementos de la matriz
print(b)                           # imprime la matriz completa
```

```python
# Numpy proporciona funciones para crear tipos específicos de matrices
# Nótese que las matrices creadas, en general, son de valores reales

a = np.zeros((2,2))   # Crea una matriz de ceros
print(a, '\n')        # la imprime con un salto de linea
                      # Para salto de linea tambiien puede usarse: print()

b = np.ones((1,2))    # Crea un vector de unos
print(b, '\n')        # Lo imprime

c = np.full((2,2), 7)  # Crea una matriz constante (todos los elementos valen lo mismo)
print(c, '\n')         # Lo imprime 

d = np.eye(3)          # Crea una matriz identidad de tamanyo 3
print(d, '\n')         # La imprime

e = np.random.random((2,2))  # Crea una matriz de elementos aleatorios entre 0 y 1
print(e, '\n')               # la imprime
```

### Ejemplos de uso de Matplotlib

Se presenta un ejemplo para representar varias curvas de funciones trigonométricas poniendo los títulos, leyendas y las etiquetas de los ejes.

```python
import matplotlib.pyplot as plt

# Se calculan los valores de las curvas de funciones seno y coseno en un rango definido
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Se pintan las curvas con la información asociada usando Matplotlib
plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.title('Gráfica de curvas seno y coseno')
plt.legend(['Seno', 'Coseno'])
plt.show()
```

### Ejemplo inicial de uso de OpenCV: carga, visualización y almacenamiento de imágenes

A continuación, se presenta una ejemplo introductorio a OpenCV donde se lee una imagen desde un fichero, se imprimen sus dimensiones, se la convierte a niveles de gris y se guarda en un fichero de salida el resultado. El ejemplo se divide en varias celdas de código, cada una dedicada a una etapa del proceso.

```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Se lee una imagen comprobando que ha sido cargada correctamente
im1 = cv2.imread('./img/Cadiz.jpg')

# Test para ver si la ha leído bien
if im1 is None:
  print('¡Problema!')
else:
  print('¡Bien!')

# Se averigua el tamaño de la imagen leída
print(im1.shape, '\n')

# También se podría leer la imagen en niveles de gris
im1_gris = cv2.imread('./img/Cadiz.jpg', 0)
print(im1_gris.shape)
```

```python
# Visualizamos la imagen en color usando Matplotlib. ¿Qué sucede?
plt.title('Imagen mal visualizada en formato BGR')
plt.imshow(im1)
```

```python
# OpenCV visualiza las imagenes en formato BGR (Blue, Green, Red), almacenando los planos de color en este orden
# Si queremos visualizarla en el formato RGB al que estamos acostumbrados, es necesario hacer algún cambio en el código.
plt.title('Imagen correctamente visualizada en formato RGB')
plt.imshow(cv2.cvtColor(im1, cv2.COLOR_BGR2RGB))
```

*Ejercicio:* Visualizar las dos imagenes anteriores de forma simultánea (en la misma fila, en vez de una debajo de otra).

```python
# Ahora es posible visualizar las imágenes anteriores de forma simultanea usando la función subplot de Matplotlib
plt.subplot(1,2,1)
plt.title('Imagen BGR')
plt.imshow(im1)

plt.subplot(1,2,2)
plt.title('Imagen RGB')
plt.imshow(cv2.cvtColor(im1, cv2.COLOR_BGR2RGB))
```

```python
# Para finalizar podemos convertir la imagen RGB a niveles en otra de niveles de gris y almacenar la imagen modificada
im2 = cv2.cvtColor(im1, cv2.COLOR_BGR2GRAY)       # Creamos una nueva imagen en niveles de gris

plt.title('Imagen en niveles de gris')              # visualizamos la imagen
plt.imshow(cv2.cvtColor(im2, cv2.COLOR_BGR2RGB))

cv2.imwrite('./img/Cadiz_Procesada.png', im2)      # la almacenamos en un archivo con un nombre
```
